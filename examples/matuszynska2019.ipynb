# Matuszyńska 2019

https://doi.org/10.1111/ppl.12962

```python
from functools import partial

from mxlpy import Scipy, Simulator, plot

from mxlbricks import get_matuszynska2019 as get_model

EPS = 1e-6

model = get_model(variant="mm-speedup")
res = (
    Simulator(model, integrator=partial(Scipy, method="Radau"))
    .simulate(100)
    # .simulate(5)
    .get_result()
    .unwrap_or_err()
)

fig, ax = plot.lines(
    res.get_variables(include_derived_variables=False, include_readouts=False)
)
ax.legend(loc="upper left", bbox_to_anchor=(1, 1), ncols=2)
plot.show()
```
